```python
import os
from PIL import Image

# Directorio donde se encuentran las imgenes
input_directory = 'descargas'
# Directorio donde se guardarn las imgenes convertidas
output_directory = 'webp'

# Crear el directorio de salida si no existe
os.makedirs(output_directory, exist_ok=True)

# Recorrer todos los archivos en el directorio de entrada
for filename in os.listdir(input_directory):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):  # Tipos de imagen a convertir
        # Ruta completa del archivo de entrada
        input_path = os.path.join(input_directory, filename)
        # Ruta completa del archivo de salida
        output_path = os.path.join(output_directory, f"{os.path.splitext(filename)[0]}.webp")
        
        # Abrir la imagen y convertirla a WebP
        with Image.open(input_path) as img:
            img.save(output_path, 'WEBP')
            print(f"Convertido: {output_path}")
```
